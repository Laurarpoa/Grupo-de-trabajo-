```
%matplotlib inline
import matplotlib.pyplot as plt

# Usar un estilo moderno
plt.style.use("seaborn-v0_8")

# Datos
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]
temperaturas = [22, 24, 27, 30, 32, 35, 34, 33, 30, 27, 24, 22]

# Crear figura
plt.figure(figsize=(10, 6))

# Graficar lÃ­nea
plt.plot(meses, temperaturas, color='red', marker='o', linewidth=2.5, markersize=8, alpha=0.85)

# Rellenar debajo de la lÃ­nea
plt.fill_between(meses, temperaturas, color='red', alpha=0.1)

# TÃ­tulo y etiquetas con estilo
plt.title("ðŸ“Š Temperatura media mensual", fontsize=20, fontweight='bold', color='darkred', pad=20)
plt.xlabel("Mes", fontsize=14, fontweight='bold', labelpad=10, color='dimgray')
plt.ylabel("Temperatura (Â°C)", fontsize=14, fontweight='bold', labelpad=10, color='dimgray')

# Ajustar ticks
plt.xticks(fontsize=12, color='black')
plt.yticks(fontsize=12, color='black')

# Activar grid
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar grÃ¡fico
plt.show()
```

```
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Estilo de Seaborn
sns.set_style("whitegrid")

# Generar datos sintÃ©ticos
np.random.seed(42)
ninos = np.random.normal(loc=130, scale=10, size=30)        # Media 130 cm, desviaciÃ³n 10
adultos = np.random.normal(loc=170, scale=8, size=30)       # Media 170 cm, desviaciÃ³n 8
adultos_mayores = np.random.normal(loc=165, scale=9, size=30)  # Media 165 cm, desviaciÃ³n 9

# Crear DataFrame
df = pd.DataFrame({
    "Altura": np.concatenate([ninos, adultos, adultos_mayores]),
    "Grupo": ["NiÃ±os"]*30 + ["Adultos"]*30 + ["Adultos Mayores"]*30
})

# Crear boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=df,
    x="Grupo",
    y="Altura",
    palette="Set2",      # Colores automÃ¡ticos
    showfliers=True      # Mostrar outliers
)

# TÃ­tulo y etiquetas
plt.title("ComparaciÃ³n de alturas por grupo", fontsize=16, fontweight='bold')
plt.xlabel("Grupo", fontsize=14, fontweight='bold')
plt.ylabel("Altura (cm)", fontsize=14, fontweight='bold')

# Mostrar grÃ¡fico
plt.show()
```

```
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo
sns.set_style("whitegrid")

# Generar datos sintÃ©ticos
np.random.seed(42)
meses = [f"Mes {i}" for i in range(1, 13)]
regiones = ["Norte", "Centro", "Sur"]

# Crear DataFrame
data = []
for mes in meses:
    for region in regiones:
        ventas = np.random.randint(1000, 5000)
        data.append([mes, region, ventas])

df = pd.DataFrame(data, columns=["Mes", "Region", "Ventas"])

# Calcular promedio mÃ³vil (agrupando por mes)
ventas_totales_mes = df.groupby("Mes")["Ventas"].sum().reset_index()
ventas_totales_mes["Promedio_Movil"] = ventas_totales_mes["Ventas"].rolling(window=3).mean()

# Crear figura
plt.figure(figsize=(10, 6))

# Barras de ventas por regiÃ³n
sns.barplot(
    data=df,
    x="Mes",
    y="Ventas",
    hue="Region",
    palette="Set2"
)

# LÃ­nea del promedio mÃ³vil (sobre ventas totales)
plt.plot(
    ventas_totales_mes["Mes"],
    ventas_totales_mes["Promedio_Movil"],
    color="black",
    marker="o",
    linewidth=2,
    label="Promedio mÃ³vil"
)

# TÃ­tulo y etiquetas
plt.title("Ventas mensuales por regiÃ³n y promedio mÃ³vil", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("Mes", fontsize=14, fontweight='bold')
plt.ylabel("Ventas", fontsize=14, fontweight='bold')

# Leyenda
plt.legend(title="Leyenda", labels=["Promedio mÃ³vil", "Norte", "Centro", "Sur"])

# Mejorar visibilidad
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar grÃ¡fico
plt.show()
```
